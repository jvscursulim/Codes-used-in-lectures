```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, plot_state_qsphere
%matplotlib inline
```

```python
qubits = QuantumRegister(2, name='qubits')
bits = ClassicalRegister(2, name='bits')

qc = QuantumCircuit(qubits,bits)
```

```python
qc.h(qubits[0])
qc.cx(qubits[0],qubits[1])
qc.barrier()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
state_vector = execute(qc,backend=backend).result().get_statevector()
plot_state_qsphere(state_vector)
```

```python
qc.x(qubits[0])
#qc.y(qubits[0])
#qc.z(qubits[0])
qc.barrier()
qc.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
state_vector = execute(qc,backend=backend).result().get_statevector()
plot_state_qsphere(state_vector)
```

```python
qc.cx(qubits[0],qubits[1])
qc.h(qubits[0])
qc.barrier()
qc.measure(qubits,bits)
qc.draw('mpl')
```

```python
qc = qc.reverse_bits()
```

```python
backend = Aer.get_backend('statevector_simulator')
state_vector = execute(qc,backend=backend).result().get_statevector()
plot_state_qsphere(state_vector)
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc,backend=backend,shots=1024)
plot_histogram(job.result().get_counts())
```
