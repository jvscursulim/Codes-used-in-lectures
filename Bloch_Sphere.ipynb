```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector, plot_state_qsphere
import numpy as np
%matplotlib inline
```

```python
qubit = QuantumRegister(1, name='qubit')

qc = QuantumCircuit(qubit)

theta = (float(input("Insira o 창ngulo theta em graus: ")))*np.pi/(180)
phi = (float(input("Insira o 창ngulo phi em graus: ")))*np.pi/(180)

alpha = complex(np.cos(theta/2))
beta = complex(np.exp(1j*phi))*np.sin(theta/2)

initial_state = np.array([alpha,beta])
```

```python
qc.initialize(initial_state,qubit)

qc.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
state_vector = execute(qc, backend=backend).result().get_statevector()
plot_bloch_multivector(state_vector)
```

```python
plot_state_qsphere(state_vector)
```
